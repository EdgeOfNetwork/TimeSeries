```python
df.groupby('month').agg(sum) 
```

걍 month로 때려버리니까,  
  
각 연도의 월 별 합계가 되어 야 하는데  09년 10년 모두 합쳐졌다??

```python
df.groupby(df.index.year).sum()
```

```python
df.groupby([df.index.year, df.index.month]).agg(sum)
```

문제를 이렇게 해결 할 수 있지만, 그다지 권장할만한 해결방법은 아니다  

https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html  


  off set들을 이용해서, 

```python
df.resample(rule='A').sum() # 매년 12월 31일을 기준으로 
```

```python
df.resample(rule='A').mean()
```

```python
df.resample(rule='A').mean()['2009']
```

```python
def first_day(sample): return sample[0]
```

```python
df.resample(rule='A').apply(first_day) #연도 샘플의 제일 첫버째 값
```
