```python
#변수 정의
target='BAD'
event ='1'

```

```python
df.dtypes
```

```python
xvars = [x for x in df.columns.tolist() if x not in [target]]

xvars
```

```python
#범주변수
cats = [x for x in df.select_dtypes('varchar').columns.tolist() if x in xvars] #categorycal의 약자
cats
```

```python
nummeric = list(set(xvars) - set(cats))
nummeric
```

```python
# 데이터 분할
cas.sampling.stratified(
    
    # 데이터 대상
    table = {'name':df, 'groupby':target},
    
    # 비율 지정
    samppct = 70, 
    
    # 분할 변수 출력 여부 결정
    partInd = True,
    
    # 출력 테이블 지정
    output = {'casout':{'name':'df_partn', 'replace':True}, 'copyVars':"ALL"}

)
```

```python
cas.CASTable('df_partn').head() #Partind : 이 변수가 1인 경우가 Training data set을 쓴다는 뜻
```

```python
train = cas.CASTable('df_partn').query("_PartInd_=1")
```

```python
train.shape
```

```python
train.columns
```

```python
# 상관관계 분석: 상호정보
train.exploreCorrelation (
    casout={'name':'corr_out', 'replace':True},
    target=target,
    inputs=xvars,
    nominals=cats
)
```

```python
cas.CASTable('corr_out').head(999).sort_values("MI",ascending=False)
```

```python
train.nmiss()
```

```python
# 결측값 처리
train.dataPreprocess.impute(
    
    # 결측값 처리 방식
    methodInterval='mean',
    methodNominal = 'mode', #mode최빈값 
    
    # 처리 변수
    inputs = xvars,
    
    # 결과 테이블
    casout={'name':'train_impute_out', 'replace':True}, 

    # 복사 변수
    copyVars = target,
    
    # 모델 저장
    code = {'casout':{'name':'impute_code', 'replace':True}} 
    
)#모델 저장을 하는 이유?? 나중에 test를 하기 위해서??? 이해를 잘 못했다
```

```python
cas.CASTable("train_impute_out").nmiss() #IMP_가 붙었따.
```

```python
cas.loadtable(
    path = 'impute_code.sashdat',
    casout = {'name':'impute_code', 'replace':True}
)
```

```python
#SCORING 매기는 작업
# 결측값 처리: 모든 데이터
cas.dataStep.runCodeTable( #DataStep으로 만든애들은 다 run code Table로 만들어주면 된다 
    # 대상
    table = 'df_partn',
    #  출력
    casout={'name':'df_partn_impute'},
    # 변수
    dropvars = xvars, #있는 모드 변수에 IMP_ 
    # 모델 테이블
    codeTable = 'impute_code'
)
```

```python
cas.CASTable("df_partn_impute").head(999)
```

```python
imp_xvars = ["IMP_" + x for x in xvars]
imp_cats = ["IMP_" + x for x in cats]
imp_nums = ["IMP_" + x for x in nummeric]
```

```python
imp_cats
```

```python
cas.loadactionset('regression')
```

```python
#
# 로지스틱 회귀 적합
#

r = cas.regression.logistic(
    # 입력 데이터
    table={'name':'df_partn_impute', 'caslib':'casuser'}, 
    partByVar={"name":"_PartInd_",
               "train":"1",
               "validate":"0"},
  
    # 변수 역할 정의
    classVars=imp_cats, #여기에 자동으로 원핫인코딩을 해서 적용해준다 

   # 모델 정의
    model={'depvar': [{'name': target, 'options': {'event': event}}],
           'link':'LOGIT',
           'effects': [{'vars': imp_xvars}]
          },
    
    
    # 변수 선택
    selection={'method':'FORWARD', 'choose': "SBC", 'select':"SBC", 'stop':"SBC"},
    
    # 저장
    store={'name':'logistic_model_astore', 'caslib':'casuser', 'replace': True}#"a"store라고 저장할떄 bin 파일로 저장된다. 그 테이블의 요 component가 bin이다 
)

r
```

```python
cas.CASTable("logistic_model_astore").head()

#이게 bin file이다 .
#이걸로 score를 매긴다 .
#pkl파일이라 생가하면 된다.
#평가 모델도 CAS TABLE 형태로 되는 것
```

```python
cas.loadactionset("astore") #analitycs store 
```

```python
# 점수 내기: ASTRE 모델 이용
cas.astore.score(
    # 점수 산출 대상 파일 지정
    table={'name':'df_partn_impute', 'caslib':'casuser', 'where':'_PartInd_=0'},

    # ATORE 모델 지정
    rstore='logistic_model_astore',

    # 복제할 변수 지정
    copyvars=[target],

    # 점수 산출 저장 테이블 지정
    casout={'name': 'logistic_score_out', 'replace': True}
)
```

```python
cas.CASTable("logistic_score_out").head(10)
```

#P_BAD1 은 event 값  
#P_BAD0 은 

#I_BAD : Into BAD 고로 BAD0으로 예측하겠따

#이것들을 종합해서 하는것이 오분류율을 계산 

```python
nonevent = '0'
r_logistic = cas.percentile.assess(
    # 평가를 위한 데이터 지정
    table='logistic_score_out',

    # 실제 목표변수 지정
    response=target,

    # 예측 변수 지정
    inputs=[{"name": 'P_' + target + event}],

    # 이벤트 값 지정
    event=event,

    # 이벤트 범주를 제외한 계산된 범주들을 목록화 하여 표시
    # pVar=['P_' + target + nonevent],
    # pEvent=[nonevent]
)
```

```python
r_logistic
```

```python
# 결과 테이블을 이용한 ROC 그래프

fig, ax = plt.subplots(dpi=100)

# 무작위 기준선 그리기
ax.plot([0, 1], [0, 1], "k--")

auc = r_logistic["ROCInfo"]["C"][0].round(3)
ax.plot(r_logistic['ROCInfo']['FPR'], r_logistic['ROCInfo']['Sensitivity'], 
       label=f'Logistic Model\n AUC: {auc}')

# 레이블 주기
ax.set_xlabel("False Positive Rate")
ax.set_ylabel("True Positive Rate")

# 제목 주기
ax.set_title("ROC Curve")

# 범례 주기
plt.legend(loc="best")

# 그림 출력
plt.show()
```
